## Format .json files to CSV
### Load a .json file

```python
#load in pokemon stat dataset
import json
 
# Opening JSON file
f = open('./pokedex.json')
 
# returns JSON object as a dictionary
pokedex_json = json.load(f)

pokedex_json
```

### convert dictionary into dataframe
#### keep relevant columns, can add back later

```python
import pandas as pd
import numpy as np

pokedex_df = pd.DataFrame.from_dict(pokedex_json)
pokedex_df


pokedex_df = pokedex_df.T
pokedex_df.info
pokedex_format_df = pokedex_df.drop(['genderRatio', 'heightm', 'weightkg', 'color', 'evos', 'tags', 'maxHP', 'requiredAbility',
                                     'canHatch', 'eggGroups', 'prevo', 'evoLevel', 'formeOrder', 'evoCondition', 'evoType', 'abilities',
                                     'gender', 'gen', 'evoItem', 'forme', 'baseForme', 'requiredItem', 'changesFrom',
                                     'evoRegion', 'baseSpecies', 'otherFormes', 'canGigantamax', 'evoMove', 'battleOnly',
                                     'requiredMove', 'requiredItems', 'cannotDynamax', 'forceTeraType'], axis = 1)
pokedex_format_df.head(10)


```

#### Split list in `types` into separate columns `Type1` and `Type2`

```python
pokedex_format_df[['Type1','Type2']] = pd.DataFrame(pokedex_format_df.types.tolist(), index = pokedex_format_df.index)
pokedex_format_df = pokedex_format_df.drop(['types'], axis=1)
pokedex_format_df.head(20)
```

#### Split list in `baseStats` into separate columns for each stat

```python
pokedex_format_df = pd.concat([pokedex_format_df.drop(['baseStats'], axis=1), pokedex_format_df['baseStats'].apply(pd.Series)], axis=1)
pokedex_format_df
```

#### Create `Total_Stats` column and rename/format

```python
#pokedex_format_df = pd.concat([pokedex_format_df.drop(['abilities'], axis=1), pokedex_format_df['abilities'].apply(pd.Series)], axis=1)
#pokedex_format_df

pokedex_format_df.rename(columns = {"num": "Pokedex_ID", "name": "Name",
                                     "hp": "HP", "atk": "Attack", "def": "Defense",
                                     "spa": "SpAtk", "spd": "SpDef", "spe": "Speed"}, inplace = True)

pokedex_format_df['Total_Stats'] = pokedex_format_df['HP'] + pokedex_format_df['Attack'] + pokedex_format_df['Defense'] + pokedex_format_df['SpAtk'] + pokedex_format_df['SpDef'] + pokedex_format_df['Speed']

pokedex_format_df.replace(to_replace = [None], value = np.nan, inplace = True)

pokedex_format_df

pokedex_format_df.to_csv('pokedex.csv', index = False)
```
